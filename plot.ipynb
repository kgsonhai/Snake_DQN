```python
import matplotlib.pyplot as plt
import pandas as pd


# Đọc dữ liệu từ file output.txt
file_path = "output.txt"
rolling_averages = []
scores = []

with open(file_path, "r") as file:
    for line in file:
        try:
            lines = line.strip().split()
            game, score, mean_score = lines[0], lines[1], lines[2]
            scores.append(float(score))
            rolling_averages.append(float(mean_score))
            

        except ValueError:
            print(ValueError)
            continue

rolling_avg_100 = pd.Series(scores).rolling(window=100, min_periods=1).mean().tolist()
rolling_avg_50 = pd.Series(scores).rolling(window=50, min_periods=1).mean().tolist()


plt.title("Deep Q Learning")
plt.xlabel("Number of Games")
plt.ylabel("Avg Scores")
plt.plot(scores)
# plt.plot(rolling_avg_50, label="Rolling 50")
# plt.plot(rolling_avg_100, label="Rolling 100")
plt.plot(rolling_averages, label="Rolling root")
plt.ylim(ymin=0)
plt.text(len(rolling_averages) - 1, rolling_averages[-1], str(rolling_averages[-1]))
plt.grid(True)
plt.show()
```
